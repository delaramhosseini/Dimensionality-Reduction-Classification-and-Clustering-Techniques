```python
import random
import pandas as pd
from sklearn.datasets import fetch_lfw_people
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=150 , resize=0.4)
x = lfw_people.data
y = lfw_people.target
```

```python
data_frame = pd.DataFrame(x)
data_frame
```

# PCA

```python
def pc(a):
    pca = PCA(n_components=a)
    data_after_pca = pd.DataFrame(pca.fit_transform(data_frame))
    return data_after_pca
```

```python
index_test = sorted(random.sample(range(0, 766), 192))
index_train = []
for i in range(0,766):
    if i not in index_test:
        index_train.append(i)
index_train = sorted(index_train)
```

```python
best_accuracy = 0
for b in range (10,80):
    data_after_pca = pc(b).to_numpy()
    data_test = [data_after_pca[i] for i in index_test]
    data_train = [data_after_pca[i] for i in index_train]
    target_test = [y[i] for i in index_test]
    target_train = [y[i] for i in index_train]
    clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(15,), random_state=1 , max_iter= 7000)
    clf.fit(data_train, target_train)
    MLPClassifier(solver='lbfgs',hidden_layer_sizes=(15,), random_state=1 , max_iter= 7000)
    pre_target = clf.predict(data_test)
    accuracy = (sum(pre_target == target_test)/192)*100
    print("count of component:" , b , "| accuracy:" , accuracy)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_pre_target = pre_target
    
```

```python
confusion_matrix(target_test, best_pre_target)
```
