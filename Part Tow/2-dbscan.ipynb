```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
```

```python
x = pandas.read_csv('first_clustering_dataset.csv')
x.info
```

```python
x.columns = ['A', 'B']

db_default = DBSCAN(eps = 0.0375, min_samples = 3).fit(x)
labels = db_default.labels_

colours = {}
colours[0] = 'r'
colours[1] = 'g'
colours[2] = 'b'
colours[-1] = 'k'
  
r = plt.scatter(x['A'], x['B'], color ='r');
g = plt.scatter(x['A'], x['B'], color ='g');
b = plt.scatter(x['A'], x['B'], color ='b');
k = plt.scatter(x['A'], x['B'], color ='k');
  

plt.figure(figsize =(9, 9))
plt.scatter(x['A'], x['B'])

plt.legend((r, g, b, k), ('Label 0', 'Label 1', 'Label 2', 'Label -1'))
  
plt.show()
```
